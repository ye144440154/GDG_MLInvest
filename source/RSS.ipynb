```python
import feedparser
import csv
from datetime import datetime
import schedule
import time

# 定義 RSS Feed 網址
rss_urls = [
    "https://finance.yahoo.com/rss/headline?s=TSLA",
    "https://www.cnbc.com/id/10000054/device/rss/rss.html"
]

# 初始化抓取次數和程序運行狀態
fetch_count = 0
program_running = True

keywords = [
    "台積電", "聯發科", "日月光", "鴻海", "大立光", "南亞科", 
    "半導體", "晶圓代工", "5G", "電動車", "Apple", "NVIDIA", 
    "AI", "AR", "自駕車", "晶片", "DRAM", "PCB", "Tesla"
]


# 定義抓取並存儲至 CSV 的函式
def fetch_and_save_rss():
    global fetch_count, program_running
    print(f"第 {fetch_count + 1} 次抓取 RSS 資料...")
    all_entries = []
    for url in rss_urls:
        try:
            feed = feedparser.parse(url)
            if feed.bozo:
                print(f"無法解析 RSS：{url}")
                continue
            for entry in feed.entries:
                # 存入標題、連結和發布時間
                published = entry.get("published", "未提供")
                title = entry.get("title", "未提供")
                link = entry.get("link","未提供")
               
                if any(keyword.lower() in title.lower() for keyword in keywords):
                    all_entries.append([title, link, published])
                
        except Exception as e:
            print(f"抓取或解析失敗：{url}, 錯誤：{e}")

    if all_entries:
        # 儲存到 CSV 文件
        file_exists = False
        try:
            with open("rss_feed_data.csv", "r", encoding="utf-8") as csvfile:
                file_exists = True
        except FileNotFoundError:
            pass

        with open("rss_feed_data.csv", "a", newline="", encoding="utf-8") as csvfile:
            writer = csv.writer(csvfile)
            if not file_exists:  # 如果文件不存在，先寫入標題
                writer.writerow(["標題", "連結", "發布時間"])
            writer.writerows(all_entries)
        print("已成功將資料儲存至 csv 文件。\n")

    fetch_count += 1
    if fetch_count >= 4:  # 當抓取次數達到 5 次時，停止程式
        print("已完成 5 次抓取，程式結束！")
        program_running = False
        
# 設定排程：每 2 秒鐘執行一次抓取
schedule.every(2).seconds.do(fetch_and_save_rss)

# 持續執行排程，最多執行 5 次
if __name__ == "__main__":
    while program_running:
        schedule.run_pending()
        time.sleep(1)
 
```
